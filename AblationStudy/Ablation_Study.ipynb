# **Ablation Study for Dataset 2**

Doing the hyperparameter tuning optmization with model as shufflenet as it yields the best results(test accuracy) for Dataset 2 among the three models. 

Tuning only the Learning Rate hyperparameter for Dataset 2

```python
# loading the losses and accuracies of different learning rates with shufflenet model for dataset 2.

shufflenet_lr1_training_loss = np.load('shufflenet_0.1_training_loss.npz')['arr_0']
shufflenet_lr1_training_accuracies = np.load('shufflenet_0.1_training_accuracies.npz')['arr_0']
shufflenet_lr1_val_loss = np.load('shufflenet_0.1_val_loss.npz')['arr_0']
shufflenet_lr1_val_accuracies = np.load('shufflenet_0.1_val_accuracies.npz')['arr_0']

shufflenet_lr2_training_loss = np.load('shufflenet_0.01_training_loss.npz')['arr_0']
shufflenet_lr2_training_accuracies = np.load('shufflenet_0.01_training_accuracies.npz')['arr_0']
shufflenet_lr2_val_loss = np.load('shufflenet_0.01_val_loss.npz')['arr_0']
shufflenet_lr2_val_accuracies = np.load('shufflenet_0.01_val_accuracies.npz')['arr_0']

shufflenet_lr3_training_loss = np.load('shufflenet_0.001_training_loss.npz')['arr_0']
shufflenet_lr3_training_accuracies = np.load('shufflenet_0.001_training_accuracies.npz')['arr_0']
shufflenet_lr3_val_loss = np.load('shufflenet_0.001_val_loss.npz')['arr_0']
shufflenet_lr3_val_accuracies = np.load('shufflenet_0.001_val_accuracies.npz')['arr_0']

shufflenet_lr4_training_loss = np.load('shufflenet_0.00171_training_loss.npz')['arr_0']
shufflenet_lr4_training_accuracies = np.load('shufflenet_0.00171_training_accuracies.npz')['arr_0']
shufflenet_lr4_val_loss = np.load('shufflenet_0.00171_val_loss.npz')['arr_0']
shufflenet_lr4_val_accuracies = np.load('shufflenet_0.00171_val_accuracies.npz')['arr_0']
```

```python
# Visualising the results of hyperparameter tuning over different learning rates for dataset 2 

num_epochs = 30
epochs = [i+1 for i in range(num_epochs)]
fig, axes = plt.subplots(2, 2, figsize=(25,20))
fig.suptitle('Hyperparameter Tuning Comparisons for Dataset 2', fontweight="bold",fontsize=20)

# Making subplots

axes[0,0].set_title('Training Accuracy Plot')
axes[0,0].plot(epochs, shufflenet_lr1_training_accuracies, '-', label = '0.1')
axes[0,0].plot(epochs, shufflenet_lr2_training_accuracies, '--', label = '0.01')
axes[0,0].plot(epochs, shufflenet_lr3_training_accuracies, '-*', label = '0.001')
axes[0,0].plot(epochs, shufflenet_lr4_training_accuracies, '-o', label = '0.00171')
axes[0,0].set_xlabel('Epochs')
axes[0,0].set_ylabel('Accuracy')
axes[0,0].legend()


axes[0,1].set_title('Training Loss Plot')
axes[0,1].plot(epochs, shufflenet_lr1_training_loss, '-', label = '0.1')
axes[0,1].plot(epochs, shufflenet_lr2_training_loss, '--', label = '0.01')
axes[0,1].plot(epochs, shufflenet_lr3_training_loss, '-*', label = '0.001')
axes[0,1].plot(epochs, shufflenet_lr4_training_loss, '-o', label = '0.00171')
axes[0,1].set_xlabel('Epochs')
axes[0,1].set_ylabel('Loss')
axes[0,1].legend()


axes[1,0].set_title('Validation Accuracy Plot')
axes[1,0].plot(epochs, shufflenet_lr1_val_accuracies, '-', label = '0.1')
axes[1,0].plot(epochs, shufflenet_lr2_val_accuracies, '--', label = '0.01')
axes[1,0].plot(epochs, shufflenet_lr3_val_accuracies, '-*', label = '0.001')
axes[1,0].plot(epochs, shufflenet_lr4_val_accuracies, '-o', label = '0.00171')
axes[1,0].set_xlabel('Epochs')
axes[1,0].set_ylabel('Accuracy')
axes[1,0].legend()


axes[1,1].set_title('Validation Loss Plot')
axes[1,1].plot(epochs, shufflenet_lr1_val_loss, '-', label = '0.1')
axes[1,1].plot(epochs, shufflenet_lr2_val_loss, '--', label = '0.01')
axes[1,1].plot(epochs, shufflenet_lr3_val_loss, '-*', label = '0.001')
axes[1,1].plot(epochs, shufflenet_lr4_val_loss, '-o', label = '0.00171')
axes[1,1].set_xlabel('Epochs')
axes[1,1].set_ylabel('Loss')
axes[1,1].legend()

plt.subplots_adjust(wspace=0.07, hspace=0.1) 
plt.savefig('tuning_comparisons_1.eps', format='eps')
```
