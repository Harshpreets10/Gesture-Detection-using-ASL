```python
import os
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

```

Preprocessing for dataset 1

```python
os.listdir('/content/asl_alphabet_train/asl_alphabet_train')
```

```python
training_dataset_path = '/content/asl_alphabet_train/asl_alphabet_train'
testing_dataset_path = '/content/asl_alphabet_test/asl_alphabet_test'
```

```python
training_transforms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()
])
```

```python
train_dataset = torchvision.datasets.ImageFolder(root=training_dataset_path, transform= training_transforms)
```

```python

from torch.utils.data import dataloader
train_loader = torch.utils.data.DataLoader(dataset= train_dataset, batch_size=32, shuffle = False)
```

```python
def get_mean_and_std(loader):
  mean = 0;
  std = 0;
  total_image_count = 0
  for images, _ in loader:
    image_count_in_a_batch = images.size(0)
    images = images.view(image_count_in_a_batch,images.size(1), -1)
    mean += images.mean(2).sum(0)
    std += images.std(2).sum(0)
    total_image_count += image_count_in_a_batch
    
  mean /= total_image_count
  std /= total_image_count

  return mean,std
```

```python
get_mean_and_std(train_loader)
```

```python
mean = [0.5188, 0.4989, 0.5142]
std = [0.2023, 0.2314, 0.2404] 

train_transforms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(degrees=15),
    transforms.RandomVerticalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(torch.Tensor(mean),torch.Tensor(std))

])


test_transforms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(torch.Tensor(mean),torch.Tensor(std))
])
```

```python
train_dataset_1 = torchvision.datasets.ImageFolder(root = training_dataset_path, transform = train_transforms)
test_dataset_1 = torchvision.datasets.ImageFolder(root = testing_dataset_path, transform = test_transforms)
```

```python
def show_transformed_images(dataset):
  loader = torch.utils.data.DataLoader(dataset,batch_size = 6,shuffle=True)
  batch = next(iter(loader))
  images,label = batch

  grid = torchvision.utils.make_grid(images,nrow = 3)
  plt.figure(figsize = (11,11))
  plt.imshow(np.transpose(grid,(1,2,0)))
  print("labels: ",label)
```

```python
show_transformed_images(train_dataset)
```
